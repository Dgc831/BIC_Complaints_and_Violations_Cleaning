```python
import pandas as pd

complaints_df = pd.read_csv('BIC_Complaints_Inquiries_20250617.csv')
violations_df = pd.read_csv('BIC_Issued_Violations_20250617.csv')
pd.set_option('display.max_columns', 42)
```

```python
nulls_complaints = complaints_df.isna().sum()
missing_pct_complaints = (complaints_df.isna().mean() * 100).round(2)
null_summary_complaints = pd.DataFrame({
    'Null Count': nulls_complaints,
    'Missing %': missing_pct_complaints,
}).sort_values('Missing %', ascending=False)

print("\nNull Summary (Complaints):")
print(null_summary_complaints)
```

```python
null_summary_complaints.to_csv("Complaints_Null_Summary.csv", index_label='Column')
```

```python
complaints_final = complaints_df.drop(columns=[
    'ACCOUNT STATUS', 'ACCOUNT ADDRESS', 
    'ACCOUNT CITY', 'ACCOUNT STATE', 'ACCOUNT ZIP', 'ACCOUNT PHONE',
    'VEHICLE MAKE', 'VEHICLE MODEL', 'VEHICLE TYPE', 'VEHICLE YEAR', 
    'VEHICLE COLOR', 'EXPORT DATE', 'Borough', 'Community Board', 'Council District',
    'Census Tract', 'BIN', 'BBL', 'NTA'])
```

```python
complaints_df.duplicated().sum()
```

```python
complaints_final.loc[:, 'DATE COMPLAINT/INQUIRY REPORTED ON'] = pd.to_datetime(
    complaints_final['DATE COMPLAINT/INQUIRY REPORTED ON'], errors='coerce'
)

complaints_final.loc[:, 'TOTAL RESPONSE TIME'] = pd.to_numeric(
    complaints_final['TOTAL RESPONSE TIME'], errors='coerce'
)

print("\nComplaints Final Data Types After Conversion:")
print(complaints_final.dtypes)
```

```python
complaints_final = complaints_df.dropna(subset=[
    'NATURE OF COMPLAINT/INQUIRY',
    'SOURCE OF COMPLAINT/INQUIRY',
    'COMPLAINT/INQUIRY STATUS'
])
```

```python
top_complaints = complaints_final['NATURE OF COMPLAINT/INQUIRY'].value_counts().head(10).index

complaints_final = complaints_final.copy()  # this avoids warning
complaints_final.loc[:, 'NATURE GROUPED'] = complaints_final['NATURE OF COMPLAINT/INQUIRY'].apply(
    lambda x: x if x in top_complaints else 'OTHER'
)

print("\nGrouped Complaint Type Counts:")
print(complaints_final['NATURE GROUPED'].value_counts())
```

```python
complaints_final.loc[:, 'NATURE OF COMPLAINT/INQUIRY'] = complaints_final['NATURE OF COMPLAINT/INQUIRY'].str.upper().str.strip()
complaints_final.loc[:, 'COMPLAINT/INQUIRY STATUS'] = complaints_final['COMPLAINT/INQUIRY STATUS'].str.upper().str.strip()

complaints_final.loc[:, 'SOURCE OF COMPLAINT/INQUIRY'] = complaints_final['SOURCE OF COMPLAINT/INQUIRY'].replace({
    'E-Mail': 'EMAIL',
    'SELF-INITATED': 'SELF-INITIATED'
}).str.upper().str.strip()


complaints_final.loc[:, 'COMPLAINT/INQUIRY BOROUGH'] = complaints_final['COMPLAINT/INQUIRY BOROUGH'].str.upper().str.strip()
complaints_final.loc[:, 'COMPLAINT/INQUIRY BOROUGH'] = complaints_final['COMPLAINT/INQUIRY BOROUGH'].replace({
    'NYC': 'Manhattan'
}).str.title().str.strip()
```

```python
complaints_final.to_csv("Cleaned_Complaints_Analysis.csv", index=False)
print("Exported Cleaned_Complaints_Analysis.csv")
```
